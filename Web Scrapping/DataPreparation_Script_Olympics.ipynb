## Extracting Data From Olympics 2021 Website For Null Values Which are Given By Athletes Name

```python
#import packages
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
```

```python
#path of webdriver
PATH = "C:\Program Files (x86)\chromedriver_win32\chromedriver.exe"

#driver path
driver = webdriver.Chrome(PATH)

#url link
driver.get("https://olympics.com/tokyo-2020/olympic-games/en/results/all-sports/athletes.htm")

```

```python
try_name_list = name_list[3165:3400]
```

```python
try_name_list
```

```python
#code for extracting data from olympics website

d = try_name_list
accuracy = pd.DataFrame(index = d,columns=['Sex', 'Age', 'Event', 'Medal'])
for x in try_name_list:
    search = driver.find_element_by_id("entries-search-name")
    search.send_keys(x)
    search.send_keys(Keys.RETURN)
    search.clear()

    try:
        element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.LINK_TEXT, x))
        )
    
        element.click()
        
    except:
        continue   
        
    #gender
    text = driver.find_elements_by_xpath("//*[@id='mainContainer']/div/div[1]/div[1]/div[2]/div[1]/div/div[2]/div/div[3]/div[1]/div[3]")
    for label in text:
        a = label.text
        accuracy["Sex"][accuracy.index == x] = a
    #age
    text = driver.find_elements_by_xpath("//*[@id='mainContainer']/div/div[1]/div[1]/div[2]/div[1]/div/div[2]/div/div[3]/div[1]/div[2]")
    for label in text:
        b = label.text
        accuracy["Age"][accuracy.index == x] = b
    #event
    text = driver.find_elements_by_xpath("//*[@id='mainContainer']/div/div[1]/div[1]/div[2]/a[1]/div/table/tbody/tr/td[2]/a")
    for label in text:
        c = label.text
        accuracy["Event"][accuracy.index == x] = c
    #medal
    text = driver.find_elements_by_xpath("//*[@id='mainContainer']/div/div[1]/div[1]/div[2]/a[1]/div/table/tbody/tr/td[4]/img")
    if len(text) > 0:
        for like in text:
            user = like.get_attribute("title")
            accuracy["Medal"][accuracy.index == x] = user
    else:
        user = " "
        accuracy["Medal"][accuracy.index == x] = user
        
        
        
        
    
    
    
    driver.back()
```
