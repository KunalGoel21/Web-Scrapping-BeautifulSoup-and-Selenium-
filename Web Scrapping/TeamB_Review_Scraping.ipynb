# Review Extraction By Selenium

```python
#import packages
from selenium import webdriver as wb 
import time
import csv
import pandas as pd
from selenium.common.exceptions import NoSuchElementException
```

```python
#creating empty lists
Name=[]
Rating=[]
Oneliner=[]
Review=[]
```

## Data Extraction Code

```python
#path for driver
webD=wb.Chrome(executable_path="C:\Program Files (x86)\chromedriver_win32\chromedriver.exe")
webD.maximize_window()

### Loop for multiple pages 
for i in range(1,501):
    #changing URl
    webD.get('https://www.amazon.in/New-Apple-iPhone-11-128GB/product-reviews/B08L8CXFZ1/ref=cm_cr_getr_d_paging_btm_prev_1?ie=UTF8&reviewerType=all_reviews&pageNumber='+str(i)+'')
    time.sleep(0.1)
    for j in range(1,11):
        try:
            name=webD.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]/div/div['+str(j)+']/div/div/div[1]/a/div[2]/span')
            Name.append(name.text)
        except NoSuchElementException:
            Name.append(None)
            
        try:
            rating=webD.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]/div/div['+str(j)+']/div/div/div[2]/a[1]/i/span').get_attribute('innerHTML')
                                           
            Rating.append(rating)
        except NoSuchElementException:
            Rating.append(None)
            
        try:
            oneliner=webD.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]/div/div['+str(j)+']/div/div/div[2]/a[2]/span').text
            Oneliner.append(oneliner)
        except NoSuchElementException:
            Oneliner.append(None)
            
        try:
            review=webD.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[1]/div/div[1]/div[5]/div[3]/div/div['+str(j)+']/div/div/div[4]/span/span')
            Review.append(review.text)
        except NoSuchElementException:
            Review.append(None)
```
